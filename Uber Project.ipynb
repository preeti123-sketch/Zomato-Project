```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('UberDataset.csv')
```

# Data Preprocessing

```python
dataset['PURPOSE'].fillna("Not",inplace = True)
```

```python
dataset.head()
```

```python
dataset['START_DATE'] = pd.to_datetime(dataset['START_DATE'], errors = 'coerce')
dataset['END_DATE'] = pd.to_datetime(dataset['END_DATE'], errors = 'coerce')
```

```python
dataset.info()
```

```python
from datetime import datetime
dataset['date'] = pd.DatetimeIndex(dataset['START_DATE']).date
```

```python
dataset.head(5)
```

```python
dataset['time'] = pd.DatetimeIndex(dataset['START_DATE']).hour
```

```python
dataset.head(5)
```

```python
dataset['day-night'] = pd.cut(x = dataset['time'],bins = [0,10,15,19,24],labels = ['Morning','Afternoon','Evining','Night'])
```

```python
dataset.head(5)
```

```python
dataset.dropna(inplace = True)
```

```python
dataset.shape
```

# Data Visualization

```python
plt.figure(figsize = (20,5))

plt.subplot(1,2,1)

sns.countplot(dataset['CATEGORY'])

plt.subplot(1,2,2)

sns.countplot(dataset['PURPOSE'])
```

```python
sns.countplot(dataset['day-night'])
```

```python
dataset['Month'] = pd.DatetimeIndex(dataset['START_DATE']).month  # START_DATE se month extract karta hai

month_label = {1.0 : 'Jan',2.0 : 'Feb',3.0 : 'March',4.0 : 'April',
               5.0 : 'May',6.0 : 'June',7.0 : 'July',8.0 : 'Aug',
               9.0 : 'Sep',10.0 : 'Oct',11.0: 'Nov',12.0 : 'Dec'}  # Months ko string name asign karta hai

dataset["Month"] = dataset.Month.map(month_label)                 # number months ko string names se replace karta hai

mon = dataset.Month.value_counts(sort = False)                    # har months ke counts calculate karta hai
```

```python
dataset.head()
```

```python
df = pd.DataFrame({
    "Month": mon.values,         # Har month ka total count.
    "VALUE COUNT": dataset.groupby("Month", sort = False)['MILES'].max()})  # Har month ke max MILES ka value.

p = sns.lineplot(data = df)       # Line plot banata hai.
p.set(xlabel="Month",ylabel="VALUECOUNT" )  # Axis labels set karta he.
```

```python
dataset['DAY'] = dataset.START_DATE.dt.weekday 

data_label = { 0:'Mon',1:'Tue',2:'Wed',3:'Thur',4:'Fri',5:'Sat',6:'Sun'}

dataset['DAY'] = dataset['DAY'].map(data_label)
```

```python
dataset.head()
```

```python
day_label = dataset.DAY.value_counts()

sns.barplot(x=day_label.index, y=day_label)
plt.xlabel('DAY')
plt.ylabel('COUNT')
```
